```python
import pandas as pd

df = pd.read_csv("./dataset_mood_smartphone.csv")
```

```python
import matplotlib.pyplot as plt
```

```python
variables_list = df['variable'].unique()
```

```python
# Initialize an empty DataFrame to store the results
stats_df = pd.DataFrame(columns=['Variable', 'Mean', 'Min', 'Max'])

# Loop through each variable to calculate mean, min, and max, and add to the DataFrame
for var in variables_list:
    values = df[df['variable'] == var]['value']
    stats = values.agg(['mean', 'min', 'max']).to_dict()  # Aggregate and convert to dictionary
    new_row = pd.DataFrame({'Variable': [var], 'Mean': [stats['mean']], 'Min': [stats['min']], 'Max': [stats['max']]})
    stats_df = pd.concat([stats_df, new_row], ignore_index=True)

stats_df
```

```python
stats_df.to_csv('Statistics_variables.csv')
```

```python
df.drop(columns="Unnamed: 0", inplace=True)
```

```python
df["time"] = pd.to_datetime(df['time'])
df["date"] = pd.to_datetime(df['time'].dt.date)
```

```python
df[df['id'] == 'AS14.01']
```

```python
# Pivot the DataFrame
df_pivoted = df.pivot_table(index=['id', 'date'], columns='variable', values='value', aggfunc='mean').reset_index()
df_pivoted
```

```python
# Drop if no mood available and sort per person
df_pivoted.dropna(subset=["mood"], inplace=True)
df_pivoted.sort_values(by=['id', 'date'], inplace=True)
```

```python
# Shift the columns to get the next day
df_pivoted['next_day'] = df_pivoted['date'] + pd.Timedelta(days=1)
df_pivoted['shifted_next_day'] = df_pivoted['date'].shift(-1)
df_pivoted['shifted_id'] = df_pivoted['id'].shift(-1)
df_pivoted['next_mood'] = df_pivoted['mood'].shift(-1)
df_pivoted = df_pivoted[((df_pivoted['shifted_next_day'] == df_pivoted['next_day']) & (df_pivoted['shifted_id'] == df_pivoted['id']))]
df_pivoted.dropna(subset="next_mood")
df_pivoted
```

```python
df_pivoted.drop(columns=["shifted_id", "shifted_next_day", "next_day"], inplace=True)
```

```python
df_pivoted = df_pivoted.fillna(0)
```

```python
df_pivoted
```

```python
df_pivoted_test = df_pivoted.copy()
df_pivoted_test['next_mood'] = df_pivoted_test['next_mood'].round().astype(int)
df_pivoted_test['mood'] = df_pivoted_test['mood'].round().astype(int)
df_pivoted_test['mood'] = df_pivoted_test['mood'].apply(lambda x: 1 if x > 5 else 0)
df_pivoted_test['next_mood'] = df_pivoted_test['next_mood'].apply(lambda x: 1 if x > 5 else 0)
```

```python
df_pivoted['next_mood']
```

```python
target = df_pivoted_test['next_mood']
X_dataset = df_pivoted_test.drop(columns=['next_mood', 'id', 'date'])
```

```python
target
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from imblearn.over_sampling import SMOTE
from collections import Counter

# Splitting the data into train, validation, and test sets
train_ratio = 0.7
val_ratio = 0.15
test_ratio = 0.15

# First, split into train+val and test sets
X_train, X_test, y_train, y_test = train_test_split(X_dataset, target, test_size=test_ratio, random_state=42)

# Apply SMOTE only on the training data
print('Original dataset shape %s' % Counter(y_train))
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)
print('Resampled dataset shape %s' % Counter(y_train_smote))

# Train a Random Forest Classifier
clf = RandomForestClassifier(random_state=42)
clf.fit(X_train_smote, y_train_smote)

# Optionally, you can also evaluate the model on the test set
y_test_pred = clf.predict(X_test)
print("Test Set Classification Report:")
print(classification_report(y_test, y_test_pred))
```
